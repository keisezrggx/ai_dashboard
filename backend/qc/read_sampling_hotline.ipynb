```python
import gspread
import pandas as pd
from google.oauth2.service_account import Credentials
```

```python
SERVICE_ACCOUNT_FILE = '../../key/Credentials.json'
SCOPES = ['https://www.googleapis.com/auth/spreadsheets',
          'https://www.googleapis.com/auth/drive']
creds = Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE, scopes=SCOPES)
client = gspread.authorize(creds)

sheet = client.open('[4] AI QC Inbound CRM Review 语音智能质检打标复审').worksheet('Sampling')
data = sheet.get_all_records()

df = pd.DataFrame(data[0:], columns=data[0])
df.to_csv('../../raw_data/sampling_hotline.csv', index=False)
```

```python
# Removing a column and apply lowercase & removing extra spaces in data
df_clean = (
    df[df['present'] == 'TRUE'].copy()
)

# Lowering column names and removing extra spaces
df_clean.columns = (
    df_clean
    .columns.str.lower()
    .str.strip()
    .str.replace(' ','_')
)

df_clean.to_csv('../../dataset_qc/sampling_hotline.csv', index=False)

df_clean
```
