```python
import re
import gspread
import pandas as pd
from google.oauth2.service_account import Credentials
```

```python
SERVICE_ACCOUNT_FILE = 'key/Credentials.json'
SCOPES = ['https://www.googleapis.com/auth/spreadsheets',
          'https://www.googleapis.com/auth/drive']
creds = Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE, scopes=SCOPES
)
client = gspread.authorize(creds)

sheet = client.open('[4] AI QC Inbound CRM Review 语音智能质检打标复审').worksheet('Sheet4')
data = sheet.get_all_records()

df = pd.DataFrame(data[1:], columns=data[0])
```

```python
df
```

```python
col = df['a']

date_pattern = r'\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}'

dates = []
uid_found = False

for val in col:
    if isinstance(val, str) and 'uid' in val:
        uid_found = True
        continue

    if isinstance(val, str) and re.match(date_pattern, val):
        if uid_found:
            dates.append(val)
            uid_found = False

result = pd.DataFrame({'Extracted Dates': dates})

result
```
