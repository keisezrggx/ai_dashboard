```python
import gspread
import pandas as pd
from google.oauth2.service_account import Credentials
```

## Pipeline A

```python
SERVICE_ACCOUNT_FILE = '../../key/credentials.json'
SCOPES = ['https://www.googleapis.com/auth/spreadsheets',
          'https://www.googleapis.com/auth/drive']

creds = Credentials.from_service_account_file(SERVICE_ACCOUNT_FILE, scopes=SCOPES)
client = gspread.authorize(creds)

sheet = client.open("Chatbot Ops Report").worksheet("Raw Data Robot Succes ratio")
data = sheet.get_all_values()

df = pd.DataFrame(data[1:], columns=data[0])
df.to_csv('../../raw_data/robot_success_ratio.csv', index=False)
```

```python
df_clean = (
    df.copy()
    .rename(columns=lambda c: c.encode('ascii', 'ignore').decode('ascii').strip())
    .replace('', pd.NA)
    .dropna(subset=['Connected to robot'], how='all')
)

df_clean
```

```python
# Removing last row
df_clean = df_clean[:-1]
```

```python
df_clean['Date'] = pd.to_datetime(df_clean['Date'], format='%d-%b-%Y', errors='coerce')

clean_list = ['Robot handle ratio',
              'Robot actual completion ratio', 
              'Robot Success ratio']

for col in clean_list:
    df_clean[col] = df_clean[col].str.replace('%','',regex=False).astype(float)

df_clean
```
