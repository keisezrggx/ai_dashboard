```python
import pandas as pd
```

```python
df_task = pd.read_csv(
    '../dataset_wellbore/包含 1500920的话术 - 工作表1.csv'
)

df_dataset = pd.read_csv(
    '../dataset_wellbore/NEW Akulaku - AI Kula Scene Classification (2025) - NEW 12.14.csv'
)
```

```python
df_dataset
```

```python
# Clean dataset
df_task = df_task.applymap(lambda x: x.strip().lower() if isinstance(x, str) else x)
df_dataset = df_dataset.applymap(lambda x: x.strip().lower() if isinstance(x, str) else x)

# clean columns names
df_dataset = df_dataset.applymap(lambda x: x.replace('\n', ' ').replace('\r', ' ') if isinstance(x, str) else x)

df_dataset.columns
```

```python
dataset_trimmed = df_dataset[['New-Code','ID\n(P1 Plan: Poin 2)', '回复\nID\nAnswer ', 'NEW - ANSWER']]
dataset_trimmed
```

```python
df_task.merge(dataset_trimmed, how='left', left_on='code', right_on='New-Code').to_csv('../dataset_wellbore/result_FAQ_id_example.csv', index=False)
```
